```python
import pandas as pd
import numpy as np
import datetime as dt
```

# Reading data

```python
file_path = 'AAPL.csv'
dataDF = pd.read_csv(file_path)
```

# Adding "day_of_week" column

```python
dataDF['Date'] = pd.to_datetime(dataDF['Date'])
dataDF['day_of_week'] = dataDF['Date'].dt.weekday_name
```
